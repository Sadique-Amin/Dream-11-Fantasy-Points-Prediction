```python
import csv

# Data provided
grounds_data = [
    {"Ground_Name": "Ahmedabad", "Ground_ID": 57851, "HomeAverage": 29.93, "HomeRPO": 8.47, "AwayAverage": 22.49, "AwayRPO": 8.05, "NeutralAverage": 28.41, "NeutralRPO": 9.21},
    {"Ground_Name": "Bengaluru", "Ground_ID": 57897, "HomeAverage": 28.43, "HomeRPO": 8.89, "AwayAverage": 27.08, "AwayRPO": 8.59, "NeutralAverage": 21.69, "NeutralRPO": 7.14},
    {"Ground_Name": "Eden Gardens", "Ground_ID": 57980, "HomeAverage": 26.66, "HomeRPO": 8.25, "AwayAverage": 24.16, "AwayRPO": 8.06, "NeutralAverage": 21.56, "NeutralRPO": 7.69},
    {"Ground_Name": "Mohali", "Ground_ID": 57991, "HomeAverage": 27.96, "HomeRPO": 8.37, "AwayAverage": 27.05, "AwayRPO": 8.32, "NeutralAverage": 23.5, "NeutralRPO": 7.72},
    {"Ground_Name": "Chennai", "Ground_ID": 58008, "HomeAverage": 27.44, "HomeRPO": 7.95, "AwayAverage": 22.93, "AwayRPO": 7.74, "NeutralAverage": 22.27, "NeutralRPO": 7.04},
    {"Ground_Name": "Delhi", "Ground_ID": 58040, "HomeAverage": 26.06, "HomeRPO": 8.19, "AwayAverage": 24.64, "AwayRPO": 7.97, "NeutralAverage": 20.11, "NeutralRPO": 6.58},
    {"Ground_Name": "Dharamsala", "Ground_ID": 58056, "HomeAverage": 25.28, "HomeRPO": 8.2, "AwayAverage": 24.62, "AwayRPO": 7.87, "NeutralAverage": 19.27, "NeutralRPO": 7.74},
    {"Ground_Name": "Hyderabad", "Ground_ID": 58142, "HomeAverage": 26.41, "HomeRPO": 8.24, "AwayAverage": 25.46, "AwayRPO": 8.12, "NeutralAverage": 25.24, "NeutralRPO": 7.99},
    {"Ground_Name": "Jaipur", "Ground_ID": 58162, "HomeAverage": 25.92, "HomeRPO": 7.75, "AwayAverage": 22.72, "AwayRPO": 7.54, "NeutralAverage": 36.0, "NeutralRPO": 9.45},
    {"Ground_Name": "Wankhede", "Ground_ID": 58324, "HomeAverage": 27.2, "HomeRPO": 8.32, "AwayAverage": 25.62, "AwayRPO": 8.23, "NeutralAverage": 32.56, "NeutralRPO": 10.22},
    {"Ground_Name": "Visakhapatnam", "Ground_ID": 58547, "HomeAverage": 21.23, "HomeRPO": 7.74, "AwayAverage": 24.24, "AwayRPO": 8.08, "NeutralAverage": 0.0, "NeutralRPO": 0.0},
    {"Ground_Name": "Pune", "Ground_ID": 545380, "HomeAverage": 23.29, "HomeRPO": 7.73, "AwayAverage": 26.31, "AwayRPO": 8.01, "NeutralAverage": 0.0, "NeutralRPO": 0.0},
    {"Ground_Name": "Guwahati", "Ground_ID": 680295, "HomeAverage": 27.96, "HomeRPO": 7.99, "AwayAverage": 26.3, "AwayRPO": 7.73, "NeutralAverage": 0.0, "NeutralRPO": 0.0},
    {"Ground_Name": "Lucknow", "Ground_ID": 1070094, "HomeAverage": 26.1, "HomeRPO": 7.56, "AwayAverage": 21.35, "AwayRPO": 7.27, "NeutralAverage": 18.95, "NeutralRPO": 6.95},
]

def calculate_batting_friendliness(ground):
    # Extract metrics
    home_avg = ground["HomeAverage"]
    home_rpo = ground["HomeRPO"]
    away_avg = ground["AwayAverage"]
    away_rpo = ground["AwayRPO"]
    neutral_avg = ground["NeutralAverage"]
    neutral_rpo = ground["NeutralRPO"]

    # Count valid components (non-zero values)
    avg_components = [home_avg, away_avg, neutral_avg]
    rpo_components = [home_rpo, away_rpo, neutral_rpo]
    valid_avg = [x for x in avg_components if x > 0]
    valid_rpo = [x for x in rpo_components if x > 0]

    # Calculate overall average and RPO
    overall_avg = sum(valid_avg) / len(valid_avg) if valid_avg else 0
    overall_rpo = sum(valid_rpo) / len(valid_rpo) if valid_rpo else 0

    # Define reference ranges
    avg_min, avg_max = 20, 30  # Batting average range
    rpo_min, rpo_max = 7, 9    # RPO range

    # Normalize batting average
    if overall_avg <= avg_min:
        avg_score = 0.0
    elif overall_avg >= avg_max:
        avg_score = 1.0
    else:
        avg_score = (overall_avg - avg_min) / (avg_max - avg_min)

    # Normalize RPO
    if overall_rpo <= rpo_min:
        rpo_score = 0.0
    elif overall_rpo >= rpo_max:
        rpo_score = 1.0
    else:
        rpo_score = (overall_rpo - rpo_min) / (rpo_max - rpo_min)

    # Final batting-friendliness score
    batting_friendliness = (avg_score + rpo_score) / 2
    return round(batting_friendliness, 3), overall_avg, overall_rpo

# Process each ground and add batting-friendliness metric
results = []
for ground in grounds_data:
    batting_friendliness, overall_avg, overall_rpo = calculate_batting_friendliness(ground)
    ground_result = {
        "Ground_Name": ground["Ground_Name"],
        "Ground_ID": ground["Ground_ID"],
        "Batting_Friendliness": batting_friendliness
    }
    results.append(ground_result)

# Write to CSV
headers = ["Ground_Name", "Ground_ID", "Batting_Friendliness", "Overall_Average", "Overall_RPO"]
with open('ground_batting_friendliness.csv', 'w', newline='') as file:
    writer = csv.DictWriter(file, fieldnames=headers)
    writer.writeheader()
    writer.writerows(results)

print("CSV file 'ground_batting_friendliness.csv' has been created successfully.")
```
