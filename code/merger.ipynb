```python
def calculateBattingFantasy(df):
    df["Fantasy"] = df["Runs"] + 4 * df["Fours"] + 6 * df["Sixes"] + 4
    df.loc[df["Runs"] >= 25, "Fantasy"] += 4
    df.loc[df["Runs"] >= 50, "Fantasy"] += 8
    df.loc[df["Runs"] >= 75, "Fantasy"] += 12
    df.loc[df["Runs"] >= 100, "Fantasy"] += 16
    df.loc[(df["Runs"] == 0) & (df["IsOut"] == 1), "Fantasy"] -= 2
    df.loc[df["SR"] > 170, "Fantasy"] += 6
    df.loc[(df["SR"] <= 170) & (df["SR"] > 150), "Fantasy"] += 4
    df.loc[(df["SR"] <= 150) & (df["SR"] > 130), "Fantasy"] += 2
    df.loc[df["SR"] < 50, "Fantasy"] -= 6
    df.loc[(df["SR"] >= 50) & (df["SR"] < 60), "Fantasy"] -= 4
    df.loc[(df["SR"] >= 60) & (df["SR"] < 70), "Fantasy"] -= 2
    return df
```

```python
def convertInnings(forWhat, value):
  if forWhat == "Consistency":
    if value < 50:
      return 1
    elif value < 100:
      return 2
    elif value < 125:
      return 3
    elif value < 150:
      return 4
    else:
      return 5
  if forWhat == "Form":
    if value < 5:
      return 1
    elif value < 10:
      return 2
    elif value < 12:
      return 3
    elif value < 15:
      return 4
    else:
      return 5
  if forWhat == "Opps":
    if value < 3:
      return 1
    elif value < 5:
       return 2
    elif value < 7:
      return 3
    elif value < 10:
      return 4
    else:
      return 5
  if forWhat == "Venue":
    if value < 2:
      return 1
    elif value < 3:
      return 2
    elif value < 4:
      return 3
    elif value < 5:
      return 4
    else:
       return 5


def convertAvg(value):
  if value < 10:
     return 1
  elif value < 20:
    return 2
  elif value < 30:
    return 3
  elif value < 40:
    return 4
  else:
    return 5


def convertSR(value):
  if value < 50:
     return 1
  elif value < 60:
    return 2
  elif value < 80:
    return 3
  elif value < 100:
    return 4
  else:
    return 5


def convertCenturies(forWhat, value):
  if forWhat == "Consistency":
    if value < 5:
      return 1
    elif value < 10:
      return 2
    elif value < 15:
      return 3
    elif value < 20:
      return 4
    else:
      return 5
  if forWhat == "Form":
    if value < 2:
      return 1
    elif value < 3:
      return 2
    elif value < 4:
      return 3
    elif value < 5:
      return 4
    else:
      return 5
  if forWhat == "Opps":
    if value <= 1:
      return 3
    elif value == 2:
      return 4
    else:
      return 5
  if forWhat == "Venue":
    if value <= 1:
      return 4
    else:
      return 5


def convertZeros(forWhat, value):
  if forWhat == "Consistency":
    if value < 5:
      return 1
    elif value < 10:
      return 2
    elif value < 15:
      return 3
    elif value < 20:
      return 4
    else:
      return 5
  if forWhat == "Form" or forWhat == "Opps":
    if value < 2:
      return 1
    elif value < 3:
      return 2
    elif value < 4:
      return 3
    elif value < 5:
      return 4
    else:
      return 5


def convertFifties(forWhat, value):
  if forWhat == "Consistency":
    if value < 10:
      return 1
    elif value < 20:
      return 2
    elif value < 30:
      return 3
    elif value < 40:
      return 4
    else:
      return 5
  if forWhat == "Form" or forWhat == "Opps":
    if value < 3:
      return 1
    elif value < 5:
      return 2
    elif value < 7:
      return 3
    elif value < 10:
      return 4
    else:
      return 5
  if forWhat == "Venue":
    if value <= 1:
      return 4
    else:
      return 5


def convertHS(value):
    if value < 25:
      return 1
    elif value < 50:
      return 2
    elif value < 100:
      return 3
    elif value < 150:
      return 4
    else:
      return 5


def convertOvers(value, forWhat):
  if forWhat == "Consistency":
    if value < 100:
      return 1
    elif value < 250:
      return 2
    elif value < 500:
      return 3
    elif value < 1000:
      return 4
    else:
      return 5
  if forWhat == "Form" or forWhat == "Opposition":
    if value < 10:
      return 1
    elif value < 25:
      return 2
    elif value < 50:
      return 3
    elif value < 100:
      return 4
    else:
      return 5
  if forWhat == "Venue":
    if value < 10:
      return 1
    elif value < 20:
      return 2
    elif value < 30:
      return 3
    elif value < 40:
      return 4
    else:
      return 5

def convertBowlingAvg(value):
    if value < 25:
      return 5
    elif value < 30:
      return 4
    elif value < 35:
      return 3
    elif value < 50:
      return 2
    else:
      return 1


def convertBowlingSR(value):
  if value < 30:
    return 5
  elif value < 40:
    return 4
  elif value < 50:
    return 3
  elif value < 60:
    return 2
  else:
    return 1

def convertBowlingFF(value, forWhat):
  if forWhat == "Consistency":
    if value < 3:
      return 3
    elif value < 5:
      return 4
    else:
      return 5
  else:
    if value < 3:
      return 4
    else:
      return 5
```

```python
def getBatPriors(df):
  df = df.copy()
  df["Average"] = 0
  df["SR"] = 0
  df["Innings"] = 0
  df["Centuries"] = 0
  df["Fifties"] = 0
  df["Zeros"] = 0
  df["Consistency"] = 0
  df["Form"] = 0
  df["Venue"] = 0
  df["Opps"] = 0

  uniquePlayers = np.unique(df["ID"])

  for player_id in uniquePlayers:
    player_df = df[df["ID"] == player_id].copy()
    for i, (index, row) in enumerate(player_df.iterrows()):
        prev_df = player_df.iloc[:i]
        num_outs = prev_df[prev_df["IsOut"] == 1].shape[0]
        runs = prev_df["Runs"].sum()
        average = runs / num_outs if num_outs != 0 else runs

        balls = prev_df["BF"].sum()
        strike_rate = (runs / balls) * 100 if balls != 0 else runs * 100

        innings = prev_df.shape[0]
        centuries = prev_df[prev_df["Runs"] >= 100].shape[0]
        fifties = prev_df[(prev_df["Runs"] >= 50) & (prev_df["Runs"] < 100)].shape[0]
        zeros = prev_df[(prev_df["Runs"] == 0) & (prev_df["IsOut"] != 0)].shape[0]

        global_index = index

        df.loc[global_index, ["Average", "SR", "Innings", "Centuries", "Fifties", "Zeros"]] = [average, strike_rate, innings, centuries, fifties, zeros]

        ## Consistency
        average = convertAvg(average)
        innings = convertInnings("Consistency", innings)
        strike_rate = convertSR(strike_rate)
        centuries = convertCenturies("Consistency", centuries)
        fifties = convertFifties("Consistency", fifties)
        zeros = convertZeros("Consistency", zeros)
        consistency = (
            0.4262 * average +
            0.2566 * innings +
            0.1510 * strike_rate +
            0.0787 * centuries +
            0.0556 * fifties -
            0.0328 * zeros
        )

        df.loc[global_index, "Consistency"] = consistency

        ## Form Calculation
        prev_5_df = prev_df.tail(5)
        num_outs_5 = prev_5_df[prev_5_df["IsOut"] == 1].shape[0]
        runs_5 = prev_5_df["Runs"].sum()
        average_5 = runs_5 / num_outs_5 if num_outs_5 != 0 else runs_5
        balls_5 = prev_5_df["BF"].sum()
        strike_rate_5 = (runs_5 / balls_5) * 100 if balls_5 != 0 else runs_5 * 100
        innings_5 = prev_5_df.shape[0]
        centuries_5 = prev_5_df[prev_5_df["Runs"] >= 100].shape[0]
        fifties_5 = prev_5_df[(prev_5_df["Runs"] >= 50) & (prev_5_df["Runs"] < 100)].shape[0]
        zeros_5 = prev_5_df[(prev_5_df["Runs"] == 0) & (prev_5_df["IsOut"] != 0)].shape[0]

        average_5 = convertAvg(average_5)
        innings_5 = convertInnings("Form", innings_5)
        strike_rate_5 = convertSR(strike_rate_5)
        centuries_5 = convertCenturies("Form", centuries_5)
        fifties_5 = convertFifties("Form", fifties_5)
        zeros_5 = convertZeros("Form", zeros_5)

        form = (
            0.4262 * average_5 +
            0.2566 * innings_5 +
            0.1510 * strike_rate_5 +
            0.0787 * centuries_5 +
            0.0556 * fifties_5 -
            0.0328 * zeros_5
        )

        df.loc[global_index, "Form"] = form

        ## Venue Effect Calculation
        Ground_Place = row.Ground
        ground_df = prev_df[prev_df["Ground"] == Ground_Place]

        num_outs_venue = ground_df[ground_df["IsOut"] == 1].shape[0]
        runs_venue = ground_df["Runs"].sum()
        average_venue = runs_venue / num_outs_venue if num_outs_venue != 0 else runs_venue

        balls_venue = ground_df["BF"].sum()
        strike_rate_venue = (runs_venue / balls_venue) * 100 if balls_venue != 0 else runs_venue * 100

        innings_venue = ground_df.shape[0]
        centuries_venue = ground_df[ground_df["Runs"] >= 100].shape[0]
        fifties_venue = ground_df[(ground_df["Runs"] >= 50) & (ground_df["Runs"] < 100)].shape[0]
        HS = ground_df["Runs"].max() if not ground_df.empty else 0

        average_venue = convertAvg(average_venue)
        innings_venue = convertInnings("Venue", innings_venue)
        strike_rate_venue = convertSR(strike_rate_venue)
        centuries_venue = convertCenturies("Venue", centuries_venue)
        fifties_venue = convertFifties("Venue", fifties_venue)
        HS = convertHS(HS)

        venue = (
            0.4262 * average_venue +
            0.2566 * innings_venue +
            0.1510 * strike_rate_venue +
            0.0787 * centuries_venue +
            0.0556 * fifties_venue -
            0.0328 * HS
        )

        df.loc[global_index, "Venue"] = venue

        ## Opposition effect calculation
        Opposition = row.Opposition
        opposition_df = prev_df[prev_df["Opposition"] == Opposition]

        num_outs_opposition = opposition_df[opposition_df["IsOut"] == 1].shape[0]
        runs_opposition = opposition_df["Runs"].sum()
        average_opposition = runs_opposition / num_outs_opposition if num_outs_opposition != 0 else runs_opposition

        balls_opposition = opposition_df["BF"].sum()
        strike_rate_opposition = (runs_opposition / balls_opposition) * 100 if balls_opposition != 0 else runs_opposition * 100

        innings_opposition = opposition_df.shape[0]
        centuries_opposition = opposition_df[opposition_df["Runs"] >= 100].shape[0]
        fifties_opposition = opposition_df[(opposition_df["Runs"] >= 50) & (opposition_df["Runs"] < 100)].shape[0]
        zeros_opposition = opposition_df[(opposition_df["Runs"] == 0) & (opposition_df["IsOut"] != 0)].shape[0]

        average_opposition = convertAvg(average_opposition)
        innings_opposition = convertInnings("Opps", innings_opposition)
        strike_rate_opposition = convertSR(strike_rate_opposition)
        centuries_opposition = convertCenturies("Opps", centuries_opposition)
        fifties_opposition = convertFifties("Opps", fifties_opposition)
        zeros_opposition = convertHS(zeros_opposition)

        opps = (
            0.4262 * average_opposition +
            0.2566 * innings_opposition +
            0.1510 * strike_rate_opposition +
            0.0787 * centuries_opposition +
            0.0556 * fifties_opposition -
            0.0328 * zeros_opposition
        )

        df.loc[global_index, "Opps"] = opps
  return df
```

```python
import re

# Dictionary mapping phrases to their distinct league names.
# Order matters: these checks are case-insensitive.
special_leagues = {
    'indian premier league': 'IPL',
    'ipl': 'IPL',
    'big bash league': 'BBL',
    'bbl': 'BBL',
    'vitality blast': 'Vitality Blast',
    'pakistan super league': 'PSL',
    'psl': 'PSL',
    'caribbean premier league': 'Caribbean Premier League',
    'mzansi super league': 'Mzansi Super League',
    'the hundred': 'The Hundred',
    'sa20': 'SA20'
}

def map_series(series_name):
    s = series_name.lower()

    # Check if series matches any of the special league names
    for key, league in special_leagues.items():
        if key in s:
            return league

    # If not one of the special leagues, then apply broader rules:

    # World events: World Cup / World Twenty20
    if re.search(r'world cup|world twenty20', s):
        return 'World Cup'

    # Bilateral series: any series with 'tour'
    elif re.search(r'\btour\b', s):
        return 'Bilateral'

    # Cup competitions: any title including 'cup'
    elif re.search(r'\bcup\b', s):
        return 'Cup'

    # Tournament, Challenge, or Tri-series events
    elif re.search(r'tri[-\s]?series|tournament|challenge', s):
        return 'Tournament'

    # If “T20” appears but none of the above apply, label it as T20 Tournament
    elif 't20' in s:
        return 'T20 Tournament'

    # Default fallback category
    else:
        return 'Other'

series_list = np.unique(batEngg["Series"])

mapped = {name: map_series(name) for name in series_list}
for name, cat in mapped.items():
    print(f"{name} -> {cat}")
```
