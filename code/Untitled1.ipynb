```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv("/content/Drumsticks - BatData (1).csv")
```

```python
df.columns
```

```python
cat_cols = ['Opposition', 'Ground', 'TossResult', 'MatchTime', 'Tournament', 'Role', 'Country', 'Batting Hand']
for col in cat_cols:
    df[col] = df[col].astype('category')
```

```python
df.drop(["IsOut"], axis = 1, inplace = True)
```

```python
df = df[(df["Fantasy"] >= 20) & (df["Fantasy"] <= 160)].copy()
df = df[df["BattingOrder"] < 4].copy()
```

```python
df
```

```python
probcols = ["SR", "Average", "Innings"]
df = df[~(df[probcols] == 0).any(axis=1)]
df
```

```python
df.columns
```

```python
import sklearn
from sklearn.metrics import accuracy_score
import matplotlib.pyplot as plt
import time
import joblib
from sklearn.model_selection import GridSearchCV
from imblearn.over_sampling import SMOTE
import xgboost as xgb
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score
from sklearn.preprocessing import LabelEncoder
from sklearn.model_selection import GridSearchCV
from sklearn.metrics import mean_squared_error
from sklearn.metrics import mean_absolute_error
from xgboost import XGBRegressor
```

```python
x_train, x_test, y_train, y_test = train_test_split(df.drop("Fantasy", axis = 1), df["Fantasy"])
```

```python
import matplotlib.pyplot as plt
from xgboost import XGBRegressor
from sklearn.metrics import mean_squared_error, mean_absolute_error

accuracyEta = []
Eta = []
plt.ion()
fig, ax = plt.subplots()
line, = ax.plot([], [], 'bo-')

# Hyperparameter ranges
n_estimators_range = range(300, 600, 200)
max_depth_values = [15]
min_child_weight_values = [5, 10, 20]
gamma_values = [5, 10, 20]
learning_rate_values = [0.01, 0.05, 0.1]
colsample_bytree_values = [0.5, 0.8, 1]
subsample_values = [0.5, 0.8, 1]

# Iterate through all parameter combinations
for eta in n_estimators_range:
    for depth in max_depth_values:
        for min_child_weight in min_child_weight_values:
            for gamma in gamma_values:
                for learning_rate in learning_rate_values:
                    for colsample in colsample_bytree_values:
                        for subsample in subsample_values:
                            print(f"Training with: eta={eta}, depth={depth}, min_child_weight={min_child_weight}, gamma={gamma}, "
                                  f"learning_rate={learning_rate}, colsample_bytree={colsample}, subsample={subsample}")

                            # Initialize XGBRegressor with current parameters
                            xgb = XGBRegressor(
                                objective='reg:squarederror',
                                eval_metric="rmse",
                                n_estimators=eta,
                                learning_rate=learning_rate,
                                max_depth=depth,
                                random_state=42,
                                enable_categorical=True,
                                device="cuda",
                                min_child_weight=min_child_weight,
                                gamma=gamma,
                                subsample=subsample,
                                colsample_bytree=colsample,
                            )

                            # Train the model
                            xgb.fit(x_train, y_train)

                            # Evaluate on validation data
                            y_pred_val = xgb.predict(x_test)
                            val_mse = mean_squared_error(y_test, y_pred_val)
                            val_mae = mean_absolute_error(y_test, y_pred_val)
                            print(f"Validation - MSE: {val_mse:.4f}, MAE: {val_mae:.4f}")

                            # Evaluate on training data
                            y_pred_train = xgb.predict(x_train)
                            train_mse = mean_squared_error(y_train, y_pred_train)
                            train_mae = mean_absolute_error(y_train, y_pred_train)
                            print(f"Train - MSE: {train_mse:.4f}, MAE: {train_mae:.4f}")
                            print("=" * 80)
```

```python
xgbREG = XGBRegressor(
            objective='reg:squarederror',
            eval_metric="rmse",
            n_estimators=300,
            learning_rate=0.05,
            max_depth=15,
            random_state=42,
            enable_categorical=True,
            device="cuda",
            min_child_weight=5,
            gamma=10,
            subsample=0.8,
            colsample_bytree=0.5,
        )
xgbREG.fit(x_train, y_train)
mean_squared_error(y_test, xgbREG.predict(x_test))
```

```python
xgbREG.save_model("XGBREG.json")
```
